```python
def fetch_processor(x):
    if x in ["Intel Core i7","Intel Core i5","Intel Core i3"]:
        return x
    else:
        sp=x.split()[0]
        if sp=="Intel":
            return "Other Intel Processor"
        elif sp=="AMD":
            return "AMD Processor"
        else:
            return "Samsung Processor"
```

```python
laptop["CPU_brand"]=laptop["CPU_model"].apply(lambda x:fetch_processor(x))
```

```python
def fetch_os(x):
    if x in ["Windows 10","Windows 7","Windows 10 S"]:
        return "Windows"
    elif x in ["macOS","Mac OS X"]:
        return "Mac"
    else:
        return "Others/No OS/Linux"
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.15,random_state=2)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
```

```python
step1 = ColumnTransformer(transformers = [
    ("col_tnf", OneHotEncoder(sparse = False, drop = "first"), [0,1,3,7,8])
], remainder = "passthrough")

step2 = LinearRegression()

pipe = Pipeline([
    ("step1", step1),
    ("step2", step2)
])
pipe.fit(x_train, y_train)
```

```python
y_pred=pipe.predict(x_test)
print("r2 score :",r2_score(y_test,y_pred))
print("mae score :",mean_absolute_error(y_test,y_pred))
```

```python
step1 = ColumnTransformer(transformers = [
    ("col_tnf", OneHotEncoder(sparse = False, drop = "first"), [0,1,3,7,8])
], remainder = "passthrough")

step2 = RandomForestRegressor(n_estimators = 100, random_state = 3, max_samples = 0.5, max_features = 0.75, max_depth = 15)

pipe = Pipeline([
    ("step1", step1),
    ("step2", step2)
])
pipe.fit(x_train, y_train)
```

```python
y_pred=pipe.predict(x_test)
print("r2 score :",r2_score(y_test,y_pred))
print("mae score :",mean_absolute_error(y_test,y_pred))
```
